```python
class Chocolate:
    def __init__(self, price: float, addition: str = None):
        if price <= 0:
            raise ValueError("Цена должна быть больше нуля.")
        self.price = price
        self.addition = addition

    def show_my_chocolate(self):
        if self.addition:
            print(f"Шоколадный батончик с добавкой: {self.addition}")
        else:
            print("Обычный шоколадный батончик")

    def __str__(self):
        return f" {self.price} рублей"

try:
    chocolate = Chocolate(100.0, "орехи")

    chocolate.show_my_chocolate().__str__()

    print(chocolate)

except ValueError as e:
    print(e)
```
